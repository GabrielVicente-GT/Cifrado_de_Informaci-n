## Librerias

```python
import math
from PIL import Image
from os import urandom
from Crypto.Cipher import DES, DES3, AES
from Crypto.Util.Padding import pad, unpad
from Crypto.Random import get_random_bytes
from cryptography.hazmat.backends import default_backend
from cryptography.hazmat.primitives.ciphers import Cipher, algorithms, modes
```

## DES

```python

def encrypt_des(data, key):
    cipher = DES.new(key, DES.MODE_ECB)
    return cipher.encrypt(pad(data, DES.block_size))


def decrypt_des(data, key):
    cipher = DES.new(key, DES.MODE_ECB)
    return unpad(cipher.decrypt(data), DES.block_size)
```

## 3DES

```python
def encrypt_3des(data, key):
    cipher = DES3.new(key, DES3.MODE_ECB)
    return cipher.encrypt(pad(data, DES3.block_size))

def decrypt_3des(data, key):
    cipher = DES3.new(key, DES3.MODE_ECB)
    return unpad(cipher.decrypt(data), DES3.block_size)
```

## AES

```python


def encrypt_aes(data, key):
    cipher = AES.new(key, AES.MODE_ECB)
    return cipher.encrypt(pad(data, AES.block_size))



def decrypt_aes(data, key):
    cipher = AES.new(key, AES.MODE_ECB)
    return unpad(cipher.decrypt(data), AES.block_size)
```

```python

def encrypt_cbc(data, key, iv):
    cipher = AES.new(key, AES.MODE_CBC, iv)
    return cipher.encrypt(pad(data, AES.block_size))

def encrypt_ecb(data, key):
    cipher = AES.new(key, AES.MODE_ECB)
    return cipher.encrypt(pad(data, AES.block_size))


def decrypt_cbc(data, key, iv):
    cipher = AES.new(key, AES.MODE_CBC, iv)
    return unpad(cipher.decrypt(data), AES.block_size)

def decrypt_ecb(data, key):
    cipher = AES.new(key, AES.MODE_ECB)
    return unpad(cipher.decrypt(data), AES.block_size)

def main():
    image_path = r"./foto1/foto1.jpeg"
    with open(image_path, 'rb') as image_file:
        image_data = image_file.read()

    aes_key = b'01234567890123456789012345678901'

    iv = get_random_bytes(AES.block_size)

    encrypted_cbc = encrypt_cbc(image_data, aes_key, iv)
    decrypted_cbc = decrypt_cbc(encrypted_cbc, aes_key, iv)

    encrypted_ecb = encrypt_ecb(image_data, aes_key)
    decrypted_ecb = decrypt_ecb(encrypted_ecb, aes_key)


    with open(r'./foto1/encrypted_cbc_foto1.jpeg', 'wb') as f:
        f.write(encrypted_cbc)

    with open(r'./foto1/decrypted_cbc_foto1.jpeg', 'wb') as f:
        f.write(decrypted_cbc)

    with open(r'./foto1/encrypted_ecb_foto1.jpeg', 'wb') as f:
        f.write(encrypted_ecb)

    with open(r'./foto1/decrypted_ecb_foto1.jpeg', 'wb') as f:
        f.write(decrypted_ecb)

if __name__ == "__main__":
    main()
```
